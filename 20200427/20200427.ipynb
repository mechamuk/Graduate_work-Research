## 가압경수로(Pressurized Water Reactor: PWR)
* PWR은 경수(Light Water, H2O)를 감속재 및 냉각재로 사용하기 때문에 경수형 원자로(Light Water Reactor: LWR)로 분류

### 원자로냉각재계통(Reactor Coolant System: RCS)
* 원자로를 담고 있는 원자로 압력용기와 이에 연결된 2, 3 또는 4개의 동일한 열전달 회로로 구성
* 배관의 부식 및 침식을 최소화하는 것
* 원자로냉각재압력경계내 페라이트 기기가 구조 건전성을 적절히 확보하고 있는지 확인하기 위해 수압시험과 예상운전과도사건을 포함한 정상운전조건시 원자로냉각재압력경계에 사용되는 압력-온도 제한치를 설정
* 안전등급 1 및 2기기에 사용되는 페라이트 재료의 파괴인성은 KEPIC MNB-2300 및 MNC-2300의 요건을 만족해야 한다.

### 원자로 압력용기(Reactor Vessel)
* 원자로냉각재계통의 가장 중요한 구성요소
* 원자로 압력용기의 재질은 저탄소강
* 원자로 내부의 냉각재와 접촉하는 표면은 부식을 방지하기 위해 오스테나이트 스테인레스 강으로 피복되어 있다.
* 원자로용기에 사용되는 주요 페라이트 재료는 최종안전성분석보고서 표 5.2-4(원자로 냉각재계통 재료 및 압력경계 기기용 용접 재료)에 제시되어 있으며, KEPIC MD 및 MN 요건을 따르고 있어 적절하다. 또한 원자로용기 상부덮개의 볼트 재료에 대해서는 10 CFR 50, Appendix G 및 USNRC Reg. Guide 1.65, “원자로용기 상부덮개 스터드의 재료 및 검사”에 관한 요건을 따르고 있어 적절하다. 원자로용기 재료의 파괴인성에 대 한 초기값은 10 CFR 50 Appendix G 및 원자력안전위원회 고시 제2014-14호의 요건을 만족하였다.
* SA508 규격 내에서도 고강도를 갖는 SA508 Gr.3 cl.2 또는 SA508 Gr.4N의 적용을 고려할 수 있다. SA508 Gr.3 cl.2 강재가 높은 강도를 갖는 대신 인성이 떨어지는 특성을 나타내는 반면, SA508 Gr.4N 강재는 Ni과 Cr 함량의 증가에 따른 경화능의 향상으로, 기존 SA508 Gr.3 강에서 나타나는 상부베이나이트 조직을 마르텐사이트/베이나이트 조직으로 변화시켜 SA508 Gr.3 강보다 향상된 강도와 인성을 가지고 있다[4,5]

### 가압기
* 원자로공급자 또는 열전달회로의 수에 무관하게 한 개가 있다
* 정상운전중 원자로냉각재계통의 압력을 일정하게 유지하고 과도상태에서 압력변화를 제어하는 기능

### 증기발생기(Steam Generator) - 작품에 적용불가

* 최근의 증기발생기에는 Alloy 690 TT 전열관을 사용하고 있으며 회사별로 다른 열처리 공정을 적용

#### 18-8스테인리스강
크롬 18%, 니켈 8%를 철에 가해서 만든 스테인리스강을 말한다.

에이틴 에이트라고도 한다. 13크롬강과 함께 가장 많이 보급된 스테인리스강이다. 13크롬강(13-8 스텐레스강)보다 값이 비싸므로 일용품으로서는 고급품이다. 이처럼 크롬 ·니켈을 많이 가하면 순철(純鐵)에서는 900~1,400℃에서만 존재하는 면심입방정(面心立方晶)이 상온까지 안정하게 되고 이것을 주체로 한 합금, 즉 오스테나이트가 되므로, 보통의 철과는 달리 강자성(强磁性)이 아니다. 탄소의 양을 0.08% 이하로 제한하는 것이 보통이지만 크롬의 탄화물이 생겨 결정입자 주변의 크롬의 농도가 내려가서 중요한 내식성(耐蝕性)을 해치므로 1,000℃ 이상으로 가열하여 이것들을 용해시켜 냉각처리한다. 탄소를 0.03% 이하로 제한한 저탄소 18-8도 있다. 내식재료로 널리 사용되고 있으며, 판재(板材) ·관재(管材) 외에 주물도 만들어진다.

# 사용후 핵연료
* 습식 저장방식
* 건식저장방식
* 사용후핵연료란 원자로의 연료로 사용된 핵연료물질 또는 기타의 방법으로 핵분열 시킨 핵연료물질을 말한다. 2015. 7. 21부터 시행된 원자력안전법 제2조제18호에서는 제35조 4항에 따라 원자력진흥위원회의 심의·의결을 통해 폐기하기로 결정한 사용후핵연료를 방사성폐기물로 정의하고 있다.
* 그러나, 사용후핵연료에는 원전연료의 원료가 되는 물질이 남아 있어 재활용이 가능하다. 또, 사용후핵연료는 원자로에서 인출된 이후에도 오랜 기간동안 높은 준위의 방사성과 열을 방출하고 임계사고의 가능성이 있다는 특성 때문에 중저준위폐기물에 비해 보다 철저한 안전관리가 요구된다. 즉, 사용후핵연료는 에너지 자원의 효율적인 이용 측면에서나 방사성폐기물의 안전관리 측면에서나 매우 중요하다.
* 1,000MWe 용량의 전형적인 경수로형 원전에서는 연간 약 20~30MTU(Metric Ton of Uranium)의 사용후핵연료가 발생되며, 1990년대 말을 기준으로 세계적으로 연간 약 10,000MTU의 사용후핵연료가 발생되고 있는 것으로 조사된 바 있다. 가압중수로형 원전인 CANDU(Canada Deuterium Uranium) 원자로에서는 경수로에 비하여 단위 출력당 더 많은 사용후핵연료가 발생되고 있다.
* 원전 운영으로 발생된 사용후핵연료는 우선 원자로 건물내 저장수조에서 냉각되며, 사용후핵연료 저장수조의 저장용량이 초과하는 경우 저장용량에 여유가 있는 다른 저장수조 또는 건식 및 습식 중간저장시설로 이송하여 저장하거나, 사용후핵연료를 재처리할 수 있다. 사용후핵연료의 처분기술은 Conditioning 기술을 제외하고는 고준위폐기물 처분기술과 유사하여 지하 500~1000m 정도의 깊이에 심지층처분하는 것이 기본이나 OECD/NEA 국가를 주축으로 심해저처분에 대해 연구를 수행중에 있다. 그러나 아직까지 전세계적으로 사용후핵연료를 직접 영구처분할 수 있는 시설은 부재한 상황이다.

## 	건식저장방식
* 금속캐스크(Metal Cask) 방식
* 콘크리트 사일로(Concrete Silo) 방식
* 횡형 콘크리트 저장모듈(Horizontal Concrete Modular Storage)방식
* 볼트 저장(Vault Storage)방식
* 수송·저장겸용(Transport and Storage) 캐스크 방식

### 금속 캐스크방식
* 사용후연료를 금속 캐스크에 담아 콘크리트 패드 위에 저장
* 캐스크구조. 몸체, 핵연료 바스켓, 뚜껑으로 구성
    * 캐스크 몸체의 외부 표면은 냉각 효율성을 높이기 위하여 여러종류의 냉각핀을 부착하기도 함

###  콘크리트 사일로 방식
* 기본적으로 금속 캐스크 방식과 개념이 유사
* 사용후연료를 콘크리트 용기에 저장하는 이 방식은 사용후연료에서 발생한 열이 콘크리트 용기와 탄소강재의 밀봉 바스켓으로 형성된 도너츠 형태의 공간에 흐르는 공기에 의해 냉각되도록 설계되었다.

### 횡형 저장모듈방식
* 사용후연료가 장전된 차폐 캐니스터를 저장부지에 설치된 콘크리트 저장모듈에 수평상태로 저장
* 차폐 캐니스터는 방사성물질이 외부로 누출되지 않도록 하며, 운반 및 저장기간중에 사용후연료를 지지 및 보호하는 역할

### 볼트저장방식
* 콘크리트 구조물내에 금속 튜브를 설치하여 튜브내에 사용후연료를 장입 밀봉하여 저장
